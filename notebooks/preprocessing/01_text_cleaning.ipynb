```python
sarcasm_headlines = pd.read_json("../../data/sarcasm-headlines.json", lines=True)
sarcasm_headlines.head()
```

```python
sarcasm_headlines.headline.str.contains(f"[{string.punctuation}]", regex=True).value_counts()
```

```python
def remove_punctuation(text):
    translator = str.maketrans("", "", string.punctuation)
    return text.translate(translator)
```

```python
sarcasm_headlines['headline'] = sarcasm_headlines['headline'].map(remove_punctuation)
```

```python
sarcasm_headlines.headline.str.contains(f"[{string.punctuation}]", regex=True).value_counts()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop = set(stopwords.words('english'))

def remove_stopwords(text):
    filtered_words = [word.lower() for word in text.split() if word.lower() not in stop]
    return " ".join(filtered_words)
```

```python
sarcasm_headlines['headline'] = sarcasm_headlines['headline'].map(remove_stopwords)
```

```python
import re
```

```python
(
    sarcasm_headlines.headline
        .str.lower()
        .str.split()
        .apply(lambda x: bool(set(x) & stop))
        .value_counts()
)
```
